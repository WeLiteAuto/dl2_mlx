```python
def normal(x, mu, sigma):
    p = 1.0 / math.sqrt(2 * math.pi  * sigma ** 2)
    return p * mx.exp(-0.5 / sigma ** 2 * (x - mu) ** 2)
```

```python
x = mx.arange(-7, 7, 0.01)
params = [(0, 1), (0, 2), (3, 1)]
fig, axs = plt.subplots(1, 1)

y = [normal(x=x, mu=mu, sigma=sigma) for mu, sigma in params]
for ii in range(3):
    axs.plot(x.tolist(), y[ii].tolist())
# len(y)
# axs.plot(x, y)
plt.show()
```
